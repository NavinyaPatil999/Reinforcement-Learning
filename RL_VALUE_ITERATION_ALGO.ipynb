# Define transitions manually

```python
states = ['S1','S2','S3','S4','S5','S6','S7','S8']
actions = ['L','R']

state_index = {s:i for i,s in enumerate(states)}

rewards = np.array([0,2,1,-1,3,-3,-7,5])

terminal_states = ['S7','S8']

# Transition Probabilities
# P[s][a] = list of (prob, next_state)

P = {}

for s in states:
    P[s] = {}
    for a in actions:
        P[s][a] = []
print("Navinya Patil C275")

# From S1
P['S1']['L'] = [(0.7,'S2'), (0.3,'S3')]
P['S1']['R'] = [(0.7,'S3'), (0.3,'S2')]

# From S2
P['S2']['L'] = [(0.7,'S4'), (0.3,'S1')]
P['S2']['R'] = [(0.7,'S5'), (0.3,'S1')]

# From S3
P['S3']['L'] = [(0.7,'S6'), (0.3,'S1')]
P['S3']['R'] = [(0.7,'S5'), (0.3,'S1')]

# From S4
P['S4']['L'] = [(1.0,'S4')]
P['S4']['R'] = [(1.0,'S4')]

# From S5
P['S5']['L'] = [(0.7,'S7'), (0.3,'S8')]
P['S5']['R'] = [(0.7,'S8'), (0.3,'S7')]

# From S6
P['S6']['L'] = [(0.7,'S3'), (0.3,'S5')]
P['S6']['R'] = [(0.7,'S5'), (0.3,'S3')]

# Terminal states
for s in terminal_states:
    P[s]['L'] = [(1.0,s)]
    P[s]['R'] = [(1.0,s)]
```

# VALUE ITERATION

```python
print("Navinya Patil C275")
def value_iteration(gamma=0.9, theta=1e-6):
    V = np.zeros(len(states))
    iteration = 0

    while True:
        delta = 0
        iteration += 1
        V_new = V.copy()

        for s in states:
            if s in terminal_states:
                continue

            action_values = []

            for a in actions:
                value = 0
                for prob, next_state in P[s][a]:
                    value += prob * (rewards[state_index[next_state]] + gamma * V[state_index[next_state]])
                action_values.append(value)

            V_new[state_index[s]] = max(action_values)
            delta = max(delta, abs(V_new[state_index[s]] - V[state_index[s]]))

        V = V_new

        if delta < theta:
            break

    # Derive policy
    policy = {}
    for s in states:
        if s in terminal_states:
            policy[s] = '-'
            continue

        action_values = {}
        for a in actions:
            value = 0
            for prob, next_state in P[s][a]:
                value += prob * (rewards[state_index[next_state]] + gamma * V[state_index[next_state]])
            action_values[a] = value

        policy[s] = max(action_values, key=action_values.get)

    return V, policy, iteration
```

# POLICY ITERATION

```python
print("Navinya Patil C275")
def policy_iteration(gamma=0.9, theta=1e-6):

    policy = {s:np.random.choice(actions) for s in states if s not in terminal_states}
    for s in terminal_states:
        policy[s] = '-'

    V = np.zeros(len(states))
    stable = False

    while not stable:

        # Policy Evaluation
        while True:
            delta = 0
            for s in states:
                if s in terminal_states:
                    continue
                a = policy[s]
                value = 0
                for prob, next_state in P[s][a]:
                    value += prob * (rewards[state_index[next_state]] + gamma * V[state_index[next_state]])
                delta = max(delta, abs(value - V[state_index[s]]))
                V[state_index[s]] = value
            if delta < theta:
                break
# Policy Improvement
        stable = True
        for s in states:
            if s in terminal_states:
                continue

            old_action = policy[s]
            action_values = {}

            for a in actions:
                value = 0
                for prob, next_state in P[s][a]:
                    value += prob * (rewards[state_index[next_state]] + gamma * V[state_index[next_state]])
                action_values[a] = value

            best_action = max(action_values, key=action_values.get)
            policy[s] = best_action

            if old_action != best_action:
                stable = False

    return V, policy
```

```python
print("1.Obtain the optimal policy for the given problem statement by iteratively by obtaining value for each iteration till the values converge")
V, policy, iterations = value_iteration(gamma=0.9)
print("Navinya Patil C275")
print("Optimal Values:")
for i,s in enumerate(states):
    print(s, round(V[i],2))

print("\nOptimal Policy:")
print(policy)
print("Iterations:", iterations)

print("Navinya Patil C275")
print("\n2.Observe and interpret the findings when discount factor for myopic agent is considered.")
V_myopic, policy_myopic, _ = value_iteration(gamma=0.3)

print("Policy with gamma=0.3:")
print(policy_myopic)
print("Navinya Patil C275")

print("\n3.Compare the policy iteration and value iteration algorithms in terms of time and computational complexity")

start = time.time()
value_iteration(gamma=0.9)
vi_time = time.time() - start

start = time.time()
policy_iteration(gamma=0.9)
pi_time = time.time() - start
print("Navinya Patil C275")
print("Value Iteration Time:", vi_time)
print("Policy Iteration Time:", pi_time)
```
