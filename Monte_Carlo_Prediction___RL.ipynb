#1. First Visit Monte Carlo – Value Table

```python
# First Visit Monte Carlo Prediction

episodes = [
    [("S1", 0), ("S2", 1), ("S3", 2)],
    [("S1", 1), ("S3", 3)],
    [("S2", 0), ("S3", 2)]
]

returns = {}
value_table = {}

for episode in episodes:

    G = 0
    visited = set()

    # calculate returns from end of episode
    for i in reversed(range(len(episode))):

        state, reward = episode[i]
        G = G + reward
        # C275 NAVINYA PATIL
        # check if state appeared earlier in episode
        if state not in [x[0] for x in episode[:i]]:

            if state not in returns:
                returns[state] = [] #Adds calculated return to the list

            returns[state].append(G)
            value_table[state] = sum(returns[state]) / len(returns[state]) #gives average return v(s)
```

```python
print("NAVINYA PATIL C275")
print("First Visit Monte Carlo Value Table")
for state in value_table:
    print(state, ":", value_table[state])
```

#2. Every Visit Monte Carlo – Value Table

```python
# Every Visit Monte Carlo Prediction

episodes = [
    [("S1", 0), ("S2", 1), ("S3", 2)],
    [("S1", 1), ("S3", 3)],
    [("S2", 0), ("S3", 2)]
]
# C275 NAVINYA PATIL
returns = {}
value_table = {}

for episode in episodes:

    G = 0

    for i in reversed(range(len(episode))):
      # C275 NAVINYA PATIL
        state, reward = episode[i]
        G = G + reward

        if state not in returns:
            returns[state] = []

        returns[state].append(G)
        value_table[state] = sum(returns[state]) / len(returns[state])
```

```python
print("Every Visit Monte Carlo Value Table")
print("C275, Navinya Patil")
for state in value_table:
    print(state, ":", value_table[state])
```
