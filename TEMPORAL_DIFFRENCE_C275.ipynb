```python
import numpy as np
import time
alpha = 0.1
gamma = 0.9

states = ['A','B','C','D','Terminal']

rewards = {('A','B'):0, ('B','C'):0, ('C','D'):0, ('D','Terminal'):1}
#C275
print("navinya")
```

```python
# Transition function C275
def next_state(s):
    if s == 'A':
        return 'B'
    if s == 'B':
        return 'C'
    if s == 'C':
        return 'D'
    if s == 'D':
        return 'Terminal'
    return 'Terminal'
```

Defining funtion to obtain value using TD(0)

```python
def TD0():
    V = {s:0 for s in states}
    start = time.time()
    for episode in range(1000):
        s = 'A'

        while s != 'Terminal':
            s_next = next_state(s)
            r = rewards.get((s,s_next),0)
            V[s] = V[s] + alpha * (r + gamma*V[s_next] - V[s])
            s = s_next

    end = time.time()
    return V, end-start
```

Defining funtion to obtain value using TD(1)

```python
def TD1():
    V = {s:0 for s in states}
    start = time.time()
    for episode in range(1000):
        episode_states = []
        episode_rewards = []
        s = 'A' #c275
        while s != 'Terminal':
            s_next = next_state(s)
            r = rewards.get((s,s_next),0)
           # V[s] = V[s] + alpha * (r + gamma*V[s_next] - V[s])
            episode_states.append(s)
            episode_rewards.append(r)
            s = s_next
        G = 0
#C275
        for t in reversed(range(len(episode_states))):
            G = gamma * G + episode_rewards[t]
            s = episode_states[t]
            V[s] = V[s] + alpha*(G - V[s])
    end = time.time()
    return V, end-start
    print("NAVINYA PATIL")
```

```python
v_td0, time_td0 = TD0()
v_td1, time_td1 = TD1()

print("TD(0) State Values")
print(v_td0)
print("navinya patil")
print("\nTD(1) State Values")
print(v_td1)
print("\nTime Taken TD(0):",time_td0)
print("Time Taken TD(1):",time_td1)
print("C275")
```
