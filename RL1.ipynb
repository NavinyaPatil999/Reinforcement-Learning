```python
import matplotlib.pyplot as plt
#C275
def render_env(env):
    frame = env.render()
    plt.imshow(frame)
    plt.axis("off")
    plt.show()
```

##A. CartPole-v1

```python
import gymnasium as gym
#C275
env = gym.make("CartPole-v1", render_mode="rgb_array")
state, info = env.reset(seed=42);
print(state)
#C275
max_steps = 50

for step in range(max_steps):
    action = env.action_space.sample()
    state, reward, terminated, truncated, info = env.step(action)
    render_env(env)

    if terminated or truncated:
        break
#C275
env.close()
```

##B. Mountain cart.

```python
import gymnasium as gym
#C275
env = gym.make("MountainCar-v0", render_mode="rgb_array") # 'env' created here
state, info = env.reset() # 'state' initialized here
#C275
max_steps = 60
#C275
for step in range(max_steps):
    action = env.action_space.sample() # 'action' is randomly sampled from 'env.action_space'
    state, reward, terminated, truncated, info = env.step(action) # 'env' uses 'action' to update 'state'
    render_env(env) # 'env' used for rendering

    if terminated or truncated:
        break
#C275
env.close() # 'env' is closed
```

#TASK 3: Identification of Reinforcement Learning Components
##1. CartPole-v1
#C275
Agent: Controller that moves the cart

Environment: CartPole simulation

State: Cart position, cart velocity, pole angle, pole angular velocity

Action: Move cart left or move cart right

Reward: +1 for every time step the pole remains balanced

Policy: Rule or strategy that decides whether to move the cart left or right based on the state

Value: Expected total future reward obtained from a given state

##2. MountainCar-v0

Agent: Controller that drives the car

Environment: MountainCar simulation

State: Car position and car velocity

Action: Push left, no push, or push right

Reward: −1 for each time step until the car reaches the hilltop

Policy: Strategy used to build momentum and reach the goal

Value: Expected cumulative reward for reaching the hilltop from a state

##3. Pendulum-v1

Agent: Controller applying torque to the pendulum

Environment: Pendulum simulation

State: Pendulum angle, angular velocity, and cosine/sine of angle

Action: Apply continuous torque to the pendulum

Reward: Negative value based on angle deviation and energy used

Policy: Strategy to apply torque so that the pendulum stays upright

Value: Expected long-term reward for keeping the pendulum balanced
